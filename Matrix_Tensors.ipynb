```python
# buat matrix X berukuran 2x3 menggunakan numpy
import numpy as py

X = py.array([[1,2,0],
              [2,1,0]])
print(X)
```

```python
# tampilkan ukuran dari matrix X
X.shape
```

```python
# tampilkan seluruh elemen dari kolom pertama dari matrix X
X[:,0]
```

```python
# tampilkan seluruh elemen dari baris kedua dari matrix X
X[1,:]
```

```python
# tampilkan seluruh elemen dari kolom pertama dari matrix X menggunakan pytorch
X[:,0]
```


*   TensorFlow



```python
import tensorflow as tf
```

```python
# buat matrix X berukuran 2x3 menggunakan tensorflow

X = tf.Variable([[1,2,0],
                [2,1,0]])
X
```

```python
# tampilkan ukuran dari matrix X menggunakan tensorflow
tf.shape(X)
```

```python
# tampilkan seluruh elemen dari kolom pertama dari matrix X menggunakan tensorflow
X[:,0]
```

```python
# tampilkan seluruh elemen dari baris kedua dari matrix X menggunakan tensorflow
X[1,:]
```

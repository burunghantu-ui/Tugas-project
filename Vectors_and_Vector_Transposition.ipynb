3. **lakukan soal nomor 2 menggunakan pytorch dan tensorflow**


*   pytorch




```python
# lakukan transpose pada vector x dan simpan pada variabel x_t menggunakan pytorch
import torch

x = torch.tensor([[1],
              [2],
              [0]])

x_t = x.t() #sintaks yang diperlukan tidak tersedia di file bapak
x_t
```

```python
# tampilkan ukuran dari vector x_t menggunnakan pytorch
x_t.shape
```

```python
# lakukan transpose pada vector y dan simpan pada variabel y_t menggunakan pytorch
y = torch.tensor([2, 1, 0])

y_2d = y.unsqueeze(0) #sintaks yang diperlukan tidak tersedia di file bapak
y_t = y_2d.t()
y_t
```

```python
# tampilkan ukuran dari vector x_t menggunakan pytorch
x_t.shape
```



*   Tensorflow



```python
import tensorflow as tf
```

```python
# lakukan transpose pada vector x dan simpan pada variabel x_t menggunakan tensorflow
x = tf.Variable([[1],
              [2],
              [0]])

x_t = tf.transpose(x)   #sintaks yang diperlukan tidak tersedia di file bapak#sintaks yang diperlukan tidak tersedia di file bapak
x_t
```

```python
# tampilkan ukuran dari vector x_t menggunakan tensorflow
x_t.shape
```

```python
# lakukan transpose pada vector y dan simpan pada variabel y_t menggunakan tensorflow
y = tf.Variable([2, 1, 0])

y_2d = tf.expand_dims(y, axis=0)  #sintaks yang diperlukan tidak tersedia di file bapak#sintaks yang diperlukan tidak tersedia di file bapak
y_t = tf.transpose(y_2d)
y_t
```
